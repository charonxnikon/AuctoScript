```python
import pandas as pd
import numpy as np
```

```python
columns = ["Player number", "Money", "Aquired stocks", "Want to buy", "Want to sell", "Price", "Number of stock"]
n_player = 6
```

```python
df_tmp = pd.DataFrame([[n_player, 10000, 2, 0, 0, 0, 135480400]], columns=columns)
df_tmp.to_csv('Player_requests.txt', index=False, sep=' ', header=False)
```

```python
prot = pd.read_csv('prot.txt', sep=' ', index_col=False, names=columns)
price = np.array(prot[prot["Player number"] == 1]["Price"])
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

columns = ["Player number", "Money", "Aquired stocks", "Want to buy", "Want to sell", "Price", "Number of stock"]

def clear():
    buy_num.delete(0, END)
    buy_cost.delete(0, END)
    sell_num.delete(0, END)
    sell_cost.delete(0, END)
    buy_num.insert(0, "0")
    buy_cost.insert(0, "0")
    sell_num.insert(0, "0")
    sell_cost.insert(0, "0")

def submit():
    global df_tmp
    df_tmp = pd.read_csv('Player_requests.txt', sep=' ', index_col=False, names=columns)
    try:
        if (buy_num.get() == "") | (buy_cost.get() == "") | (sell_num.get() == "") | (sell_cost.get() == ""):
            raise ValueError
        last = df_tmp.tail(1)
        num_buy = int(buy_num.get())
        cost_buy = int(buy_cost.get())
        num_sell = int(sell_num.get())
        cost_sell = int(sell_cost.get())
        if (num_buy < 0) | (cost_buy < 0) | (num_sell < 0) | (cost_sell < 0):
            raise ValueError
        if (int(last["Aquired stocks"]) - num_sell < 0) | (int(last["Number of stock"]) - num_buy < 0) | (
            int(last["Money"]) - num_buy*cost_buy < 0) | ((np.abs(cost_sell/price[len(df_tmp)] - 1) > 0.3) & (
            cost_sell != 0)) | ((np.abs(cost_buy/price[len(df_tmp)] - 1) > 0.3) & (cost_buy != 0)):
            raise RuntimeError
        df_tmp = df_tmp.append(pd.DataFrame([[n_player, int(0.97*(int(last["Money"]) + num_sell*cost_sell - num_buy*cost_buy)), 
                                              int(last["Aquired stocks"]) - num_sell + num_buy, num_buy, num_sell, cost_buy,
                                              int(last["Number of stock"]) - num_buy]],
                                            columns=columns), ignore_index=True)
        df_tmp.to_csv('Player_requests.txt', index=False, sep=' ', header=False)
        current_num_label.config(text="Текущее число акций: {}".format(int(df_tmp.tail(1)["Aquired stocks"])))
        current_cost_label.config(text=f"Текущая цена: {int(price[len(df_tmp)])}")
        current_money_label.config(text="Текущие деньги: {}".format(int(df_tmp.tail(1)["Money"])))
    except ValueError:
        messagebox.showinfo("GUI Python", "Проверьте, что все поля заполнены и заполнены неотрицательными числовыми данными")
    except RuntimeError:
        messagebox.showinfo("GUI Python", "Проверьте, что введенное число акций согласуется со здравым смыслом")
    clear()

root = Tk()
root.title("Моделирование рынков")


current_num_label = Label(text="Текущее число акций: {}".format(int(df_tmp.tail(1)["Aquired stocks"])))
current_cost_label = Label(text=f"Текущая цена: {int(price[len(df_tmp)])}")
current_money_label = Label(text="Текущие деньги: {}".format(int(df_tmp.tail(1)["Money"])))


buy_num_label = Label(text="Число акций для покупки:")
buy_cost_label = Label(text="Цена акций для покупки:")
sell_num_label = Label(text="Число акций для продажи:")
sell_cost_label = Label(text="Цена акций для продажи:")

current_num_label.grid(row=0, column=0, sticky="w")
current_cost_label.grid(row=0, column=1, sticky="w")
current_money_label.grid(row=0, column=2, sticky="w")

buy_num_label.grid(row=1, column=0, sticky="w")
buy_cost_label.grid(row=2, column=0, sticky="w")
sell_num_label.grid(row=3, column=0, sticky="w")
sell_cost_label.grid(row=4, column=0, sticky="w")

buy_num = Entry()
buy_cost = Entry()
sell_num = Entry()
sell_cost = Entry()
 
buy_num.grid(row=1,column=1, padx=5, pady=5)
buy_cost.grid(row=2,column=1, padx=5, pady=5)
sell_num.grid(row=3,column=1, padx=5, pady=5)
sell_cost.grid(row=4,column=1, padx=5, pady=5)
 
buy_num.insert(0, "0")
buy_cost.insert(0, "0")
sell_num.insert(0, "0")
sell_cost.insert(0, "0")

display_button = Button(text="Submit", command=submit)

clear_button = Button(text="Quit", command=root.destroy)
 
display_button.grid(row=5, column=0, padx=5, pady=5, sticky="e")
clear_button.grid(row=5, column=1, padx=5, pady=5, sticky="e")
 
root.mainloop()
```
